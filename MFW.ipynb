```python
import os
import collections

songs = collections.defaultdict(set)
total_words = collections.defaultdict(int)
song_names = {}
for f in sorted(os.listdir('songs')):
    if f[0] == '.':
        continue
    
    for i, l in enumerate(open('songs/' + f, 'r').readlines()):
        if i == 0:
            song_names[f.split(".")[0]] = l.strip()
            continue
            
        if i < 2:
            continue
        
        arr = l.strip().split(" ")
        
        for word in arr:
            if word != '':
                total_words[word] += 1
            
print(total_words)
```

```python
count = [0]*200
for k in total_words.keys():
    if type(count[total_words[k]]) == list:
        count[total_words[k]].append(k.lower())
    else:
        count[total_words[k]] = [k.lower()]
print(count)
```

```python
def countOccurences(str, word): 
    a = str.split(" ")
    count = 0
    for i in range(0, len(a)):
        if (word == a[i]): 
            count = count + 1      
    return count
```

```python
pts = []
for f in sorted(os.listdir('songs')):
    text = open('songs/' + f, 'r').read()
    andcoord = countOccurences(text, "our")
    thecoord = countOccurences(text, 'my')
    pts.append((andcoord, thecoord))
print(pts)
```

```python
import numpy as np
import matplotlib.pyplot as plt
pts = np.array(pts)
plt.scatter(pts[:, 0], pts[:, 1])
plt.show()
```

```python
from nltk.corpus import wordnet as wn
import string

def countType(text):
    a = text.split(" ")
    countn, countv, counts, counta = 0, 0, 0, 0
    for i in range(0, len(a)):
        
        translator = str.maketrans('', '', string.punctuation)
        a[i] = a[i].translate(translator)
        typeof = wn.synsets(a[i])
        if typeof == []:
            continue
        print(wn.synsets(a[i])[0].pos())
        if typeof[0].pos() == 'n':
            countn += 1
        elif typeof[0].pos() == 'v':
            countv += 1
        elif typeof[0].pos() == 's' or typeof[0].pos() == 'r':
            counts += 1
        elif typeof[0].pos() == 'a':
            counta += 1
    return (countn, countv, countv)
```

```python
pts = []
for f in sorted(os.listdir('songs')):
    text = open('songs/' + f, 'r').read()
    n, v, a = countType(text)
    pts.append([n, v, a])
pts
```

```python
pts = np.array(pts)
y = pts[:, 0]
z = pts[:, 1]
k = pts[:, 2]
n = sorted(os.listdir('songs'))

from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')

plt.scatter(z, y)

for i, txt in enumerate(n):
    plt.annotate(txt, (z[i], y[i]))
```
